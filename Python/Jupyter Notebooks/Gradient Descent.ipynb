```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import scipy
```

$ F(X,Y) = \sin(\frac{1}{2}x^2\ -\ \frac{1}{4}y^2\ +\ 3)\cos(2x\ +\ 1\ -\ e^y)$

```python
f = lambda xy: np.sin((1/2)*(xy[0]**2) - (1/4)*(xy[1]**2) + 3)*np.cos(2*xy[0] + 1 - np.e**xy[1])
n = 100
X = np.linspace(-3,3,n)
Y = np.linspace(-3,3,n)

Z = np.zeros((n,n))

for ix,x in enumerate(X):
    for iy,y in enumerate(Y):
        Z[ix,iy] = f([x,y])
        
T = np.random.rand(2)*4 - 2
delta = 0.001
lr = 0.001
grad = np.zeros(2)

plt.plot(T[0],T[1],'o',c='w')

for _i in range(10000):
    for i,v in enumerate(T):
        _T = np.copy(T)
        _T[i] = _T[i] + delta      # A la variable correspondiente le a√±ado delta
        deriv=(f(_T)-f(T))/delta   # Calculo la derivada PARCIAL con la variable que estoy calculando   
        grad[i] = deriv            # Actualizo el valor del gradiente
    
    T = T - lr * grad
    if _i%1000 == 0:
        print(T,f(T))
    if _i%100 == 0:
        plt.plot(T[0],T[1],'.',c='r')
        

plt.contourf(X,Y,Z,10)
plt.colorbar()
plt.plot(T[0],T[1],'o',c='cyan')
plt.show()
```
