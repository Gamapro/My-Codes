## Diferencias Divididas

```python
def DiferenciasDivididas( x = [1.0], y = [0.0], show = True):
# Este metodo crea, y regresa la funcion aproximada
    print("\nDiferenciasDivididas\n")
    di = dict()
    adj = [y]
    n = len(y)
    for i in range(1,n):
        aux = []
        for j in range((n-i)):
            # print(i+j,j,end='  ')
            val = (adj[i-1][j+1] - adj[i-1][j])/( x[i+j] - x[j] )
            aux.append(val)
        for j in range(n-i,n):
            aux.append("----------")
        adj.append(aux)
    
    sol = []
    
    for i in range(len(adj)):
        di["Index "+str(i)] = []
        for j in range(len(adj[i])):
            di["Index "+str(i)].append(str(adj[i][j]))
            if j == 0: sol.append(adj[i][j]) # print(adj[i][j])
    if show:
        file = pd.DataFrame(di)
        file.style.hide_index()
        display(file)
        print("Vector solucion:", sol)
# Crear funcion aproximada
    f = ""
    for i,val in enumerate(sol):
        if i > 0: f+= " + "
        f+="("+str(val)+")"
        for j in range(i):
            f+="*(x-("+str(x[j])+"))"
        
    print("\nFuncion:",f)
    # sdadas
    f = sp.lambdify(sp.symbols('x'),f)
    return f
```

## Secante

```python
def Secante(x0=1, x1=2, eps=1e-9, n = 50, f = "x^2-1", show = True, step = 5):

    if f == "":
        f = input("Escribe tu funcion: ")
    x = sp.symbols('x')
    print("\nMetodo Secante\n")
    print('Funcion Original:',f)
    f = sp.lambdify(x,f)
    
    di = {'Iteración': [0], 'X_i':[x0], 'X_i+1':[x1], 'F(x_i)': [f(x1)],'Err_abs':[abs(f(x1))],'Err_relativo':["100%"]}

    # Iteracion
    for i in range(n):
        x2 = x1 - f(x1) * (x1 - x0)/( f(x1) - f(x0))
        err_abs = abs(f(x2))
        err_rel = abs((x2-x1)/x2) * 100
        
        if((i+1)%step == 0):
            di["Iteración"].append(i+1)
            di["F(x_i)"].append(str(f(x2)))
            di["X_i"].append(str(x1))
            di["X_i+1"].append(str(x2))
            di["Err_relativo"].append(str(err_rel)+"%")
            di["Err_abs"].append(err_abs)
        if abs(x2-x1) < eps:
            break
        x0 , x1 = x1, x2
       
    if show:
        file = pd.DataFrame(di)
        file.style.hide_index()
        display(file)
    
    print("Solucion:", x2)
    
    return x2
```

## Newton Raphson

```python
def Newton_Raphson(x0=1, eps=1e-9, n = 100, f = "x^2-1", show = True, step = 5):

    if f == "":
        f = input("Escribe tu funcion: ")
    x = sp.symbols('x')
    df = sp.diff(f)
    print("\nNewton_Raphson\n")
    print('Funcion Original:',f,'  Derivada:',df)
    f = sp.lambdify(x,f)
    df = sp.lambdify(x,df)
    
    di = {'Iteración': [0], 'X_i':[x0],'F(x_i)': [f(x0)],'Err_abs':[abs(f(x0))],'Err_relativo':["100%"]}

    # Iteracion
    for i in range(n):
        x1 = x0 - f(x0)/df(x0)
        err_abs = abs(f(x1))
        err_rel = abs((x0-x1)/x1) * 100
        if((i+1)%step == 0):
            di["Iteración"].append(i+1)
            di["F(x_i)"].append(str(f(x1)))
            di["X_i"].append(str(x1))
            di["Err_relativo"].append(str(err_rel)+"%")
            di["Err_abs"].append(err_abs)
        if abs(x1-x0) < eps:
            break
        x0 = x1
       
    if show:
        file = pd.DataFrame(di)
        file.style.hide_index()
        display(file)
    
    print("Solucion:", x1)
    
    return x1
```

## Bisection

```python
def Bisection(a=0, b=1 , eps=1e-9, n = 100, f = "", show = True, step = 5):
    x0, i, ini = a, 0, a
    if f == "":
        f = input("Escribe tu funcion: ")
    if a == 0: x0 = b
    x = sp.symbols('x')
    sign = lambda x: math.copysign(1, x)
    print("\nBisection\n")
    print('Funcion:',f)
    f = sp.lambdify(x,f)
    di = {'Iteración': [0], 'Left':[a],'Right': [b],'X_i':[x0],'F(x_i)': [f(x0)],'Err_abs':[abs(f(x0))],'Err_relativo':["100%"]}
    if show: print("f(a):",f(a)," ---  f(b):",f(b))
    err_abs = abs(f(x0))
    err_rel = 100
    while i < n and err_rel > eps:
        x1 = a + (b-a)/2
        if f(a) * f(x1) > 0: a = x1
        else: b = x1
        i+=1
        if(i%step == 0) and show:
            err_abs = abs(f(x1))
            err_rel = abs((x0-x1)/x1) * 100
            di["Iteración"].append(i)
            di["Left"].append(a)
            di["Right"].append(b)
            di["F(x_i)"].append(str(f(x1)))
            di["X_i"].append(x1)
            di["Err_relativo"].append(str(err_rel)+"%")
            di["Err_abs"].append(err_abs)
        x0 = x1

    if show:
        file = pd.DataFrame(di)
        file.style.hide_index()
        display(file)
        
    print("Solucion:", x1)
    return x1
```

## Polinomio de Lagrange

```python
def Lagrange( x = [1.0], y = [1.0], show = True):
# Este metodo crea, y regresa la funcion aproximada
    print("\nLagrange\n")
    n = len(x)
    di = {"X_i":[],"Y_i":[],"L_i":[]}
# Crear valores de L_i
    L = []
    for i in range(n):
        l = ""
        for j in range(n):
            if i != j:
                if l != "": l+=" * "
                l+="(x - "+str(x[j])+")/("+str(x[i])+" - "+str(x[j])+")"
        L.append(l)
        
# Crear funcion aproximada
    f = ""
    for i,val in enumerate(y):
        l = L[i]
        di["X_i"].append(x[i])
        di["Y_i"].append(val)
        di["L_i"].append(l)
        if i > 0: f+= " + "
        f+="("+str(l)+")*"
        f+="("+str(val)+")"
        
    if show:
        file = pd.DataFrame(di)
        file.style.hide_index()
        display(file)
        
    print("\nFuncion:",f)
    f = sp.parse_expr(f)
    f2 = f
    print("\nFuncion Simplificada:",f)
    f = sp.lambdify(sp.symbols('x'),f)
    return f,f2
```

## Polinomio de Hermite

```python
def Hermite( x = [1.0], y = [1.0], dy = [1.0], show = True):
# Este metodo crea, y regresa la funcion aproximada
    print("\nHermite\n")
    di = dict()
    aux = []
    aux2 = []
    for i in range(len(x)):
        aux2.extend([x[i],x[i]])
    aux3 = []
    for i in range(len(y)):
        aux3.extend([y[i],y[i]])

    for i in range(len(x)*2 - 1):
        if i%2 == 0:
            aux.append( dy[int(i/2)] )
        else:
            aux.append((aux3[i+1] - aux3[i])/(aux2[i+1]-aux2[i]))
            
    adj = [aux]
    n = len(aux)
    x = aux2
    for i in range(1,n):
        aux = []
        for j in range((n-i)):
            val = (adj[i-1][j+1] - adj[i-1][j])/( x[i+j+1] - x[j] )
            aux.append(val)
        for j in range(n-i,n):
            aux.append("----------")
        adj.append(aux)
    
    aux2 = []
    for i in range(len(y)):
        aux2.extend([dy[i],"----------"])
    sol = [y[0]]
    aux = [x,aux3,aux2]
    aux.extend(adj)
    adj = aux
    
    for i in range(len(adj)):
        while len(adj[i]) < len(adj[0]):
            adj[i].append("----------")
    
    di["X"] = x
    di["F(x)"] = aux3
    di["F'(x)"] = aux2
    for i in range(3,len(adj)):
        di["Diferencia "+str(i-2)] = []
        for j in range(len(adj[i])):
            di["Diferencia "+str(i-2)].append(str(adj[i][j]))
            if j == 0: sol.append(adj[i][j]) # print(adj[i][j])
            
    if show:
        file = pd.DataFrame(di)
        file.style.hide_index()
        display(file)
        print("Vector solucion:", sol)
# Crear funcion aproximada
    f = ""
    for i,val in enumerate(sol):
        if i > 0: f+= " + "
        f+="("+str(val)+")"
        for j in range(i):
            f+="*(x-("+str(x[j])+"))"
    print("\nFuncion:",f)
    f = sp.parse_expr(f)
    f2 = f
    print("\nFuncion Simplificada:",f)
    f = sp.lambdify(sp.symbols('x'),f)
    return f,f2
```

## Punto Fijo

```python
def PuntoFijo(g = "exp(-x)", f = "exp(-x)-x", x0 = 1, eps=1e-9, n = 100, show = True, step = 5):
    print("\nPunto Fijo\n")
    print('Funcion Original:',f)
    print('Funcion Auxiliar:',g)
    x = sp.symbols('x')
    f = sp.lambdify(x,f)
    g = sp.lambdify(x,g)
    di = {'Iteración': [0], 'X_i':[x0],'G(x_i)': [g(x0)],'Err_abs':[abs(g(x0))],'Err_relativo':["100%"]}

    # Iteracion
    for i in range(n):
        x1 = g(x0)
        err_abs = abs(g(x1))
        err_rel = abs((x0-x1)/x1) * 100
        if((i+1)%step == 0):
            di["Iteración"].append(i+1)
            di["G(x_i)"].append(str(g(x1)))
            di["X_i"].append(str(x1))
            di["Err_relativo"].append(str(err_rel)+"%")
            di["Err_abs"].append(err_abs)
        if abs(x1-x0) < eps:
            break
        x0 = x1
       
    if show:
        file = pd.DataFrame(di)
        file.style.hide_index()
        display(file)
    
    print("Solucion:", x1)
```

## Gauss - Seidel

```python
def GaussSeidel(A=np.array([[6,1],[3,1]]),B=np.array([8,-4]),eps=1e-9,rep=200):  
    n,m = np.shape(A)
    X = np.zeros((n,1), dtype = float)
    it, err = 0, 2.0
    difs = np.ones( np.shape(X) , dtype = float)
    while err > eps and it < rep:
        it+=1
        for i in range(n):
            suma = 0
            for j in range(n):
                if j != i:
                    suma += A[i][j] * X[j]
            x_1 = (B[i] - suma) / A[i][i]
            difs[i] = np.abs(x_1 - X[i])
            X[i] = x_1
        err = np.max(difs)
    
    if it >= rep:
        print(" El sistema no tiene solucion. ")
        return 0
    else:
        X = np.around(X, decimals=8).flatten()
        print("Aprox:",np.around(np.dot(A,X).reshape(np.shape(B)), decimals=6))
        print("Real: ",B)
        print("X = ", X)
        return X
```

## Fadev - Leverrier

```python
def FadevLeverrier(A=np.array([[6,1],[3,1]])):  
    A = np.array(A)
    assert A.shape[0] == A.shape[1], "La matriz debe de ser cuadrada"
    n = A.shape[0]
    I = np.identity(n)
    a = A
    b = A
    sol = [1]
    for i in range(1,n+1):
        tr = a.trace()/i
        b = a - tr*I
        sol.append(-tr)
        a = A @ b
    return sol
```

## Euler

```python
def Euler(f, a, b, n, y0):
    if type(f) in [str]: 
        f = sp.lambdify([sp.symbols('y'),sp.symbols('t')],f)
    h = (b-a)/n
    t = [a]
    y = [y0]
    for i in range(n):
        t.append(t[i]+h)
        y.append(y[i]+ h * f(y[i],t[i]) )
    return list(zip(t, y))
```

## Runge Kutta K4

```python
def RungeKuttaK4(f, a, b, n, y0):
    if type(f) in [str]: 
        f = sp.lambdify([sp.symbols('y'),sp.symbols('t')],f)
    h = (b-a)/n
    t = [a]
    y = [y0]
    for i in range(n):
        k1 = h*f(y[i], t[i])
        k2 = h*f(y[i] + (k1/2) , t[i] + (h/2))
        k3 = h*f(y[i] + (k2/2) , t[i] + (h/2))
        k4 = h*f(y[i] + k3 , t[i] + h)         
        t.append(t[i]+h)
        y.append(y[i] + (k1 + 2*k2 + 2*k3 + k4)/6 )
    return list(zip(t, y))
```
